```python
# load file
# sort theo date

import pandas as pd
import numpy as np
path1 = 'prices.csv'
path2 = 'sales.csv'
df1 = pd.read_csv(path1, parse_dates=['updated_at']).sort_values('updated_at')
df2 = pd.read_csv(path2, parse_dates=['ordered_at']).sort_values('ordered_at')
```

```python
df1.head()
```

```python
df2.head()
```

```python
# merge 2 bảng theo datetime (match thời điểm order với thời điểm update giá sớm hơn gần nhất)

df = pd.merge_asof(df2, df1,
                   left_on='ordered_at',
                   right_on='updated_at',
                   by='product_id')
df
```

```python
df = df.fillna(method='bfill')
df
```

```python
# add column price 

df['price'] = np.where(df['ordered_at'] >= df['updated_at'],df['new_price'],df['old_price'])
df
```

```python
# add column revenue = new price*quantity

df['sales'] = df['price']*df['quantity_ordered']
df
```

```python
# group by product và price

df_grouped = df.groupby(by=['product_id','price']).sum()
df_grouped['sales']
```

```python
# Total revenue của mỗi product_id

df.groupby('product_id').agg({'sales':sum}) 
```

# 1. Data Correction

## 1.1 Drop NA

```python
# tạo bảng df_dropna_all bằng cách loại bỏ tất cả giá trị null trong bảng và in ra info
df_dropna_all = transactions.dropna()
df_dropna_all.info()
```

```python
# tạo bảng df_dropna_online_order bằng cách loại bỏ giá trị null trong cột online_order và in ra info
df_dropna_online_order = transactions.dropna(subset='online_order')
df_dropna_online_order.info()
```

## 1.2 Fill NA

```python
# fill tất cả NA bằng 0, lưu vào bảng df_fillna_0 và in ra info
df_fillna_0 = transactions.fillna(0)
df_dropna_all.info()
```

```python
# fill NA bằng ffill, lưu vào bảng df_fillna_ffill và in ra info
df_fillna_ffill = transactions.fillna(method="ffill")
df_fillna_ffill.info()
```

```python
# fill NA bằng bfill, lưu vào bảng df_fillna_bfill và in ra info
df_fillna_bfill = transactions.fillna(method="bfill")
df_fillna_bfill.info()
```

## 1.3. Assess Categorical Data

```python
# load sheet Demographic và hiển thị top head
demographic = pd.read_excel('KPMG.xlsx', sheet_name = 'CustomerDemographic')
demographic.head()
```

```python
# in ra info bảng 
demographic.info()
```

```python
# describe các giá trị object
demographic.describe(include='object')
```

```python
# Remove all NA row
demographic = demographic.dropna()
```

```python
# In ra các giá trị unique của cột gender
print(demographic['gender'].unique())
```

```python
# Quy chuẩn cột Gender về 2 nhóm Male và Female, Unisex và in ra giá trị unique mới
demographic['gender'] = demographic['gender'].replace({'F': 'Female', 'U': 'Unisex', 'Femal': 'Female', 'M': 'Male'})
print(demographic['gender'].unique())
```

```python
# Tính tuổi của mỗi khách hàng và lưu vào cột "Age" và in ra min age, max age
from datetime import datetime, date
def age(born):
    today = pd.to_datetime('2017-12-31')
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))  
demographic['age'] = demographic['DOB'].apply(age)
print(demographic['age'].min())
print(demographic['age'].max())
```

```python
# phân nhóm khách hàng dưới 36 tuổi là nhóm "Young", Từ 36 - 55 là "Middle", trên 55 là Older
# lưu vào cột "Age_Group" bằng cách sử dụng hàm cut
demographic['age_group']= pd.cut(demographic['age'], [0,36,55,90], labels=["Young", "Middle", "Older"])
demographic['age_group']
```

# 2. Data Completeness

```python
#join data sale vs demographic và in ra những trường hợp missing demographic info
join = transactions.merge(demographic, on='customer_id', how='left', indicator=True)
list = join.loc[join['_merge'] == 'left_only', 'customer_id']
final_list = join[join['customer_id'].isin(list)]
final_list
```

# 3. Data Aggregation

```python
# tính tổng khách hàng bằng hàm nunique
total_customer = transactions['customer_id'].nunique()
total_customer
```

```python
# Tính số khách khách hàng mỗi tháng
# hint 1: tạo column year_month bằng hàm dt.strftime
transactions['year_month'] = transactions['transaction_date'].dt.strftime('%Y/%m')
transactions['year_month']
# hint 2: group by year_month và nunique
transactions_1 = transactions.groupby('year_month',as_index=False)['customer_id'].nunique()
transactions_1
```

```python
# Tính Gross margin của mỗi tháng. Gross margin = list_price - standard_cost
# hint 1: tạo column Gross_Margin
transactions['gross_margin'] = (transactions['list_price']-transactions['standard_cost'])/transactions['list_price']
# hint 2: group by year_month và sử dụng hàm agg
transactions_2 = transactions.groupby('year_month',as_index=False)['gross_margin'].agg('mean')
transactions_2
```

```python
# Tính số lượng order và doanh thu group by online_order và order_status 
# hint 1: group by online_order, order_status và sử dụng hàm agg
transactions_3 = transactions.groupby(['online_order', 'order_status']).agg({'transaction_id':'count','list_price':'sum'})
transactions_3
```

# 4. Data Visualization

```python
# Sử dụng seaborn và vẽ ra 5 loại chart khác nhau (column, line, box,...) với đầy đủ tên chart, tên các trục và chú thích.
# Đưa ra message thông qua từng chart vừa vẽ

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# revenue và gross_margin theo tháng: column chart+kde

sales_demo = transactions.merge(demographic,on='customer_id')
sales_demo['year_month'] = sales_demo['transaction_date'].dt.strftime('%Y/%m')

revenue = transactions.groupby('year_month',as_index=False).agg({'list_price':'sum','gross_margin':'mean'})
revenue
```

```python
sns.set_style('ticks')
sns.despine(left=True)

fig,ax1 = plt.subplots(figsize=(15,10))
ax2 = ax1.twinx()

sns.barplot(x='year_month', y='list_price', data = revenue, color="purple",alpha=0.5,ax=ax1) 

sns.lineplot(x='year_month', y='gross_margin', data = revenue, color="yellow", ax=ax2)

ax2.set(ylabel = 'Gross Margin', ylim=(0.4,0.5))

ax1.bar_label(ax1.containers[0])
ax1.set(xlabel='Month', ylabel = 'Sales', title = 'Sales and Gross Margins by Months')
fig.autofmt_xdate(rotation=75)

plt.show()

# Comment: Doanh thu và lợi nhuận trong năm 2017 duy trì tương đối ổn định và ít có sự biến động lớn, có 3 tháng vượt mức $1.9M là 5,8,10. 
# Biên lợi nhuận biến động không đáng kể, giảm sâu nhất vào tháng 5/2017 (43,3%) tăng trưởng cao nhất vào tháng 8/2017 (45.5%).
```

```python
# phân bố độ tuổi khách hàng theo giới tính: histogram/kde

sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(10, 10))
sns.despine(f)

sns.histplot(
    sales_demo,
    x="age", hue="gender",
    multiple="stack",
    palette="light:m_r",
    linewidth=1,
    kde=True)

ax.set(xlabel = 'Age', ylabel = '', title='Customer demographics')
ax.set_xlim(10,80)
ax.set_xticks([10,15,20,25,30,35,40,45,50,55,60,65,70,75,80])
plt.show()

# Comment: Phần lớn đối tượng khách hàng mua xe đạp có độ tuổi trong khoảng 40-50 ở cả 3 giới tính, đặc biệt là độ tuổi 44-45
# Tỷ lệ nam/nữ tương đối đều, lượng KH nữ nhỉnh hơn một chút
```

```python
# số đơn đặt hàng theo hình thức mua hàng: count plot

fig, ax = plt.subplots()

sns.countplot(data=transactions, x="online_order", palette="tab10", alpha=.6, hue = 'product_size',ax=ax)

sns.despine(fig)
ax.set(xlabel="", ylabel="", title="Total transactions by Order method")
ax.set_xticklabels(['Offline','Online'])
ax.legend(title = 'Product Size')

# Comment: tỷ lệ số đơn hàng thực hiện online và offline là tương đương nhau, đúng với cả từng nhóm kích thước sản phẩm
```

```python
# lượng hàng bán/doanh thu của từng brand: bar plot ngang
brand = transactions.groupby('brand',as_index=False)['list_price'].sum().sort_values('list_price',ascending=False)
brand
```

```python
sns.set_style('ticks')

b = sns.catplot(
    data=brand, kind="bar",
    x="list_price", y="brand",
    palette="husl", alpha=.6, height=5)

b.despine(left=True)
b.fig.suptitle("Sales by Brands", y=1.1)
b.set_axis_labels("Sales", "")

# Comment: 3 hãng có doanh thu bán xe cao nhất là Solex, WeareA2B và Giant Bicycles (trên $4M trong năm 2017)
```

```python
# lượng bán mỗi nhóm sản phẩm: bar plot (chia thành từng brand)
product = transactions.groupby(['product_line','brand'],as_index=False)['transaction_id'].count()
product.set_index('brand')
```

```python
sns.set_style('whitegrid')

l = sns.catplot(
    data=product, kind="bar",
    x="product_line", y = 'transaction_id', hue="brand",
    palette="Paired", height=7)

l.fig.suptitle("Total transactions by Product Lines", y=1.1)
l.set_axis_labels("", "")

# Comment: Dòng xe đạp Standard chiếm ưu thế vượt trội về số lượng bán, tiếp đến là dòng xe Road. Dòng Touring và Mountain bán được khá ít
# Trong dòng xe Standard bán chạy nhất là xe của hãng Solex (3500 đơn), tiếp đến là hãng Giant và Weare (khoảng 2500 đơn)
```

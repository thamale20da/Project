```python
import datetime
import pytz

def convert_datetime_timezone(dt, tz1, tz2):
    tz1 = pytz.timezone(tz1)
    tz2 = pytz.timezone(tz2)

    dt = datetime.datetime.strptime(dt,"%Y-%m-%d %H:%M:%S")
    dt = tz1.localize(dt)
    dt = dt.astimezone(tz2)
    dt = dt.strftime("%Y-%m-%d %H:%M:%S")

    return dt
```

```python
def pre_processing_nw(data):
  # Create new columns with next event

  df1 = data.copy()
  df1 =df1[(df1.day_diff.notnull())&(df1.country.isin(list_country))]

  df1['next_event'] = df1.sort_values(by=['user','event_timestamp'], ascending=True).groupby(['user'])['event_name'].shift(-1)
  df1['session_number_1'] = df1.sort_values(by=['user','event_timestamp'], ascending=True).groupby(['user'])['session_number'].shift(-1)
  df1['level_1'] = df1.sort_values(by=['user','event_timestamp'], ascending=True).groupby(['user'])['level'].shift(-1)

  df1['win_1'] = df1.sort_values(by=['user','event_timestamp'], ascending=True).groupby(['user'])['status'].shift(-1)
  df1['day_diff_1'] = df1.sort_values(by=['user','event_timestamp'], ascending=True).groupby(['user'])['day_diff'].shift(-1)

  # Cleanning
  df1 = df1[(df1.event_name.isin(['game_start','challenge_start']))&(df1.next_event.isin(['game_end','challenge_end']))]
  df1 = df1[~((df1.event_name == 'game_start')&(df1.next_event == 'challenge_end'))]
  df1 = df1[~((df1.event_name == 'challenge_start')&(df1.next_event == 'game_end'))]

  # Adjust and check

  df_row_nw = df1[(df1.day_diff <= 7)&(df1.win_1 == True)]
  df_row_nw = df_row_nw[~(df_row_nw.level != df_row_nw.level_1)]
  df_row_nw = df_row_nw[~(df_row_nw.session_number != df_row_nw.session_number_1)]

  # Create game order

  df_row_nw['row'] = df_row_nw.groupby(['user'])['event_timestamp'].rank(method='first') 

  # Create funnel data
  df_funnel_nw = df_row_nw.groupby(['version','network_name','country','row'])[['user']].agg({'user':'nunique'})
  df_funnel_nw.reset_index(inplace = True)
  return df_row_nw,df_funnel_nw
```
